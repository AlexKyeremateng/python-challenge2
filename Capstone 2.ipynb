```python
import pandas as pd
student_data=r"C:\Data Analytics Course\Capstone Projects\Capstone 2\students_complete.csv"
df2=pd.read_csv(student_data)
df2
```

```python
df2.isnull().sum()
```

```python
combined_data = pd.merge(df1, df2, on='school_name', how='left')
print(combined_data)
```

```python
# District Summary Calculations
total_schools = combined_data["school_name"].nunique()
total_students = combined_data["student_name"].count()
total_budget = df1["budget"].sum()
average_math_score = combined_data["math_score"].mean()
average_reading_score = combined_data["reading_score"].mean()
```

```python
# % Passing Calculations

passing_math = combined_data[combined_data["math_score"] >= 70].count()["student_name"]
passing_reading = combined_data[combined_data["reading_score"] >= 70].count()["student_name"]
overall_passing = combined_data[(combined_data["math_score"] >= 70) & (combined_data["reading_score"] >= 70)].count()["student_name"]

# Percentage calculations
percent_passing_math = (passing_math / total_students) * 100
percent_passing_reading = (passing_reading / total_students) * 100
percent_overall_passing = (overall_passing / total_students) * 100
```

```python
# Create the District Summary DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]
})

district_summary
```

```python
# Group data by each school
school_group = combined_data.groupby(["school_name"])
```

```python
# Perform the calculations per school
school_type = df1.set_index("school_name")["type"]
total_students_per_school = school_group["student_name"].count()
total_budget_per_school = school_group["budget"].mean()
per_student_budget = total_budget_per_school / total_students_per_school
average_math_score_per_school = school_group["math_score"].mean()
average_reading_score_per_school = school_group["reading_score"].mean()
```

```python
# % Passing per school
passing_math_per_school = school_group.apply(lambda x: (x["math_score"] >= 70).mean()) * 100
passing_reading_per_school = school_group.apply(lambda x: (x["reading_score"] >= 70).mean()) * 100
overall_passing_per_school = school_group.apply(lambda x: ((x["math_score"] >= 70) & (x["reading_score"] >= 70)).mean()) * 100
```

```python
# Create School Summary DataFrame
per_school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students_per_school,
    "Total School Budget": total_budget_per_school,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score_per_school,
    "Average Reading Score": average_reading_score_per_school,
    "% Passing Math": passing_math_per_school,
    "% Passing Reading": passing_reading_per_school,
    "% Overall Passing": overall_passing_per_school
})

per_school_summary.head()
```

```python
# Sort schools by % Overall Passing
top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False).head(5)
top_schools
```

```python
# Sort schools by % Overall Passing in ascending order
bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True).head(5)
bottom_schools
```

```python
# Math Scores by Grade
math_scores_by_grade = combined_data.pivot_table(values='math_score', index='school_name', columns='grade', aggfunc='mean')
math_scores_by_grade
```

```python
# Reading Scores by Grade
reading_scores_by_grade = combined_data.pivot_table(values='reading_score', index='school_name', columns='grade', aggfunc='mean')
reading_scores_by_grade
```

```python
# Binning by Spending
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"], bins=spending_bins, labels=spending_labels)
```

```python
# Grouping by Spending
spending_summary = per_school_summary.groupby("Spending Ranges (Per Student)").sum()
spending_summary
```

```python
# Binning by School Size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], bins=size_bins, labels=size_labels)
```

```python
# Grouping by School Size
size_summary = per_school_summary.groupby("School Size").max()
size_summary
```

```python
# Group by School Type
type_summary = per_school_summary.groupby("School Type").max()
type_summary
```
